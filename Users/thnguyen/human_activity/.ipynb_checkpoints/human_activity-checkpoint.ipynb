```python
def print_features(headers=[]):
    features = {}
    for var in headers:
        k = var.split("-")[0]
        if k.endswith('Mag'):
            k = k[0:-3]
        if k not in features.keys():
            features[k] = 1
        else:
            features[k] = features[k]+1
    count = 0
    for k,v in features.items():
        count = count+1
        print(count, ":", k, "\t\t", v)
def print_measures(headers=[]):
    measures = {}
    for var in headers:
        k = var.split("-")
        if len(k) == 2:
            measures[k[1]] = True
        elif len(k) ==3:
            measures[k[1]+k[2]]=True
    print(measures.keys())
```

```python
headers = data[data.columns[0:-2]].columns.values.tolist()
print_features(headers)
print_measures(headers)
```

We will remove parentheses, comma, and hyphen, typo BodyBody in the header name

```python
def fix_name(names):
    old_new_names = {}
    for name in names:
        newname = name
        if "BodyBody" in name:
            newname = newname.replace("BodyBody", "Body")
        if "(" in newname:
            newname = newname.replace("(", "")
            newname = newname.replace(")", "")
        if "," in name:
            newname = newname.replace(",","_")
        if "-" in name:
            newname = newname.replace("-","_")
        old_new_names[name] = newname 
    return old_new_names
```

```python
X = data.drop(['subject','Activity'], axis=1)
X_test = data_test.iloc[:,0:561]
old_new_names = fix_name(X.columns.values.tolist())
X.rename(columns=old_new_names, inplace=True)
X_test.rename(columns=old_new_names, inplace=True)
data.rename(columns=old_new_names, inplace=True)
data_test.rename(columns=old_new_names, inplace=True)
print_features(X_test.columns.values.tolist())  
```

Confusion matrix

```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
from sklearn.metrics import confusion_matrix
cfs = confusion_matrix(data_test.Activity, pred)
from sklearn import preprocessing
labelEncoder= preprocessing.LabelEncoder()
labelEncoder.fit(data_test.Activity)
labelEncoder.classes_

class_names = labelEncoder.classes_
plot_confusion_matrix(cfs, classes=class_names,
                      title=' Confusion Matrix, without normalization')
```

The model can clasify correctly between static and dynamic activities. However, it cannot classify 2 static activities: sitting and standing and f-score is low 0.88, classification of 3 dynamic activities: walking, walking upstair, and walking downstair have f-score just arround 0.90.
