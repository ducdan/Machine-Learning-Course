## Prediction

```python
import numpy as np
from sklearn import datasets, linear_model
```

```python
df.columns
```

```python
df_prediction = df[['Population', 'Property\ncrime']]
```

```python
df_prediction_newCols = ['Population', 'PropertyCrime']
```

```python
df_prediction.columns = df_prediction_newCols
```

```python
df_prediction.head()
```

```python
df_prediction.dtypes
```

```python
df_prediction.isnull().any()
```

```python
df_prediction = df_prediction.fillna(method='ffill')
```

```python
x = df_prediction.Population.values.reshape(df_prediction.shape[0],1)
y = df_prediction.PropertyCrime.values
```

```python
regr = linear_model.LinearRegression(normalize = True)
model = regr.fit(x, y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y,  color='blue', marker="v")
plt.plot(x, model.predict(x), color='red')
plt.xticks(())
plt.yticks(())
plt.xlabel("Population")
plt.ylabel("PropertyCrime")
plt.show()
```

```python
model.coef_
```

```python
model.intercept_
```

## Cross Validation

```python
from sklearn.model_selection import cross_val_predict
```

```python
predicted = cross_val_predict(regr, x, y)
```

```python
plt.scatter(y, predicted,  color='blue', marker="v")
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red')
plt.xticks(())
plt.yticks(())
plt.xlabel("Measured")
plt.ylabel("Predicted")
plt.show()
```
